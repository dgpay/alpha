```
library(igraph)

files <- list.files(path="/content/", pattern="*.edges", full.names=TRUE, recursive=FALSE)
lapply(files, function(x) {
    t <- read.table(x, header=TRUE) # load file
    # apply function
    out <- function(t)
    # write to file
    write.table(out, "path/to/output", sep="\t", quote=FALSE, row.names=FALSE, col.names=TRUE)

    
    # print(tab)
    links <- data.frame(
        source=c(t[1]),
        target=c(t[2])
        )

    # create the network object
    network <- graph_from_data_frame(d=links, directed=T) 
    # print_all(network)
    arr<-alpha.centrality(network,alpha = 0.7)
    a<-max(arr) #nilai alpha terbesar
    indx<-match(c(a),arr) #cari index ke berapa
    IdNode=V(network) #ini idNode nya dari setiap buletan ego
    print(a)
    print(indx)
    print(IdNode)
    V(network)[indx] #ini vertex yang kuatnya


})


# ===========inputan batas=========================================
tab<-read.table ("/content/12831.edges" , header = TRUE)
# print(tab)
links <- data.frame(
    source=c(tab[1]),
    target=c(tab[2])
    )

# create the network object
network <- graph_from_data_frame(d=links, directed=T) 
# print_all(network)
plot(network)
arr<-alpha.centrality(network,alpha = 0.5)
a<-max(arr) #nilai alpha terbesar
indx<-match(c(a),arr) #cari index ke berapa
IdNode=V(network) #ini idNode nya dari setiap buletan ego

print(indx)
print(IdNode)
V(network)[indx] #ini vertex yang kuatnya

#==============================Nanti======================================

sprintf("lalala %i",IdNode[indx])



tab<-read.table ("/content/12831.circles" , header = FALSE)
circle_name <-tab[1,2]
tab1<-read.table ("/content/12831.feat" , header = FALSE)
net<- as.matrix(tab)
tab2<-read.table ("/content/12831.featnames" , header = FALSE)
net1<- as.matrix(tab2)
# print(net[1,])
# print(net[,1])
IdxFeat<-match(1260231,net)
# for (i in net){
#   if (net[IdxFeat,i] == 1) {
#     print(net1[i,1])
#   }
# }

```
